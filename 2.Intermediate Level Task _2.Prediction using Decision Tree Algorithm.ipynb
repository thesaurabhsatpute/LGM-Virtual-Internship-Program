**Importing Required Libraries**

```python
import numpy as np
import pandas as pd
import sklearn.metrics as sm
import seaborn as sns
import matplotlib.pyplot as mt
%matplotlib inline

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import plot_tree
from sklearn.metrics import accuracy_score
from matplotlib.pyplot import rcParams


import warnings
warnings.filterwarnings("ignore")
```

**Dataset reading**

```python
dat = pd.read_csv('/content/Iris.csv')
dat
```

```python
dat.describe()
```

**Checking Null Values**

```python
dat.isna().sum()
```

```python
dat.describe()
```

**Data Visualization & Cor-relation Matrix**

```python
sns.pairplot(dat, hue='Species')
```

```python
sns.heatmap(dat.corr(),annot=True)
```

**Data Preprocessing**

```python
X=dat.iloc[:,[0,1,2,3]]
y=dat.Species
y
dat.shape
```

```python
X.shape
```

```python
y.shape
```

**Train,Test,Split**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train.shape
```

```python
X_test.shape
```

**Decision Tree**

```python
clf = DecisionTreeClassifier(criterion='entropy')
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred_train = clf.predict(X_train)
y_pred_test = clf.predict(X_test)
```

```python
pd.DataFrame(dat.Species).value_counts()
```

```python
accuracy_score(y_train, y_pred_train)
```

```python
accuracy_score(y_test, y_pred_test)
```

```python
rcParams['figure.figsize'] = 80, 50
plot_tree(clf,precision=5,rounded=True,filled=True);
```
